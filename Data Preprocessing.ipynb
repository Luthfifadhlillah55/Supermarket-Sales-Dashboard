## Dataset Explanation

The growth of supermarkets in most populated cities are increasing and market competitions are also high. The dataset is one of the historical sales of supermarket company which has recorded in 3 different branches for 3 months data. Predictive data analytics methods are easy to apply with this dataset.

Attribute information

- Invoice id: Computer generated sales slip invoice identification number
- Branch: Branch of supercenter (3 branches are available identified by A, B and C).
- City: Location of supercenters
- Customer type: Type of customers, recorded by Members for customers using member card and Normal for without member card.
- Gender: Gender type of customer
- Product line: General item categorization groups - Electronic accessories, Fashion accessories, Food and beverages, Health and beauty, Home and lifestyle, Sports and travel
- Unit price: Price of each product in dolar
- Quantity: Number of products purchased by customer
- Tax: 5% tax fee for customer buying
- Total: Total price including tax
- Date: Date of purchase (Record available from January 2019 to March 2019)
- Time: Purchase time (10am to 9pm)
- Payment: Payment used by customer for purchase (3 methods are available â€“ Cash, Credit card and Ewallet)
- COGS: Cost of goods sold
- Gross margin percentage: Gross margin percentage
- Gross income: Gross income
- Rating: Customer stratification rating on their overall shopping experience (On a scale of 1 to 10)

## Data Cleaning 

### Check and Remove Column

```python
# check duplicated all columns
data.duplicated().sum()
```

There is no duplicate column from the datasets

```python
data.head(2)
```

```python
# check is two columns have equal
data["Tax 5%"].equals(data["gross income"])
```

Because the **Tax 5%** and **gross income** columns have the same value, a drop will be made to the **Tax 5%** column and only the **gross income** column will be used.

```python
data.drop('Tax 5%', axis=1, inplace=True)
```

```python
data.info()
```

```python
data.head()
```

The **Tax 5%** column has been removed.

### Handling Data Type

**Date** column and **Time** column represent time but the data type is still a float, so the data type will be changed to DateTime

```python
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M').dt.hour
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
data.head(2)
```

### Rename Columns

```python
data = data.rename(columns={'Customer type':'Customer Type',
                        'Product line': 'Product Category',
                        'Unit price':'Unit Price',
                        'cogs':'COGS',
                        'gross margin percentage':'GMP',
                        'gross income':'Gross Income',
                        'Time':'Hour'})
```

```python
data.head(2)
```

### Add Columns 
Added an **Hour Range** column by category that represents the **Hour** column

```python
data.head(3)
```

```python
buying_range = [] #list kosong

for i, k in data.iterrows():
    if k['Hour'] <= 11: 
        buying_range.append('9 ~ 11')
    elif k['Hour'] <= 14: 
        buying_range.append('12 ~ 14')
    elif k['Hour'] <= 17: 
        buying_range.append('15 ~ 17')
    elif k['Hour'] <= 20: 
        buying_range.append('18 ~ 20')
    elif k['Hour'] <= 23: 
        buying_range.append('21 ~ 23')
data['Hour Range'] = buying_range
```

```python
data.head(3)
```

**Hour Range** column has been added to the dataset
